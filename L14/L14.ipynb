```python
import mrob
import numpy as np
```

```python
graph = mrob.FGraph()
#help(graph.add_node_pose_3d)
T = mrob.geometry.SE3(np.random.randn(6))
print('initial random pose = \n',T.T())
n1 = graph.add_node_pose_3d(T)
print('node = ', n1)

# anchor factor
W_0 = np.identity(6)
# help(graph.add_factor_1pose_3d)
graph.add_factor_1pose_3d(mrob.geometry.SE3(),n1,W_0)

# add a landmark observation
W = np.identity(3)
l1 = graph.add_node_landmark_3d(np.zeros(3))
obs = np.array([1,0,0])
graph.add_factor_1pose_1landmark_3d(obs,n1,l1,W)
print(graph.get_estimated_state())

print('Initial chi2 = ' , graph.chi2())
with mrob.ostream_redirect():
    graph.solve(mrob.LM)
    print('chi2 = ' , graph.chi2())
    graph.print(True)
    pass
```
